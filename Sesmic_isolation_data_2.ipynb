```python

import numpy as np
LQ=np.percentile(data['Horizontal amplitude of driven mirror (cm)'],25)
UQ=np.percentile(data['Horizontal amplitude of driven mirror (cm)'],75)
IQR=UQ-LQ
x=UQ+(1.5*IQR)
y=LQ-(1.5*IQR)
idx=np.where((data['Horizontal amplitude of driven mirror (cm)']<x)&(data['Horizontal amplitude of driven mirror (cm)']>y))[0]
print(idx)
#print(data['Horizontal amplitude of driven mirror (cm)'][idx)
plt.plot(data['Frequency (Hz)'][idx],data['Horizontal amplitude of driven mirror (cm)'][idx],color='red',linestyle='-')
plt.grid()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude of driven filtered')
plt.savefig('plot9b.png')
```

Ploting both raw and filterd ratios

```python
plt.plot(data['Frequency (Hz)'],data['Raw'],color='red',linestyle='-')
plt.plot(data['Frequency (Hz)'],data['Filtered'],color='blue',linestyle='-')
plt.grid()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Ratio')
plt.savefig('plot7b.png')
```

```python

import numpy as np
LQ=np.percentile(data['Horizontal amplitude of driven mirror (cm)'],25)
UQ=np.percentile(data['Horizontal amplitude of driven mirror (cm)'],75)
IQR=UQ-LQ
x=UQ+(1.5*IQR)
y=LQ-(1.5*IQR)
idx=np.where((data['Horizontal amplitude of driven mirror (cm)']<x)&(data['Horizontal amplitude of driven mirror (cm)']>y))[0]
print(idx)
#print(data['Horizontal amplitude of driven mirror (cm)'][idx)
plt.plot(data['Frequency (Hz)'][idx],data['Horizontal amplitude of driven mirror (cm)'][idx],color='red',linestyle='-',label='Filtered')
plt.plot(data['Frequency (Hz)'],data['Horizontal amplitude of driven mirror (cm)'],color='blue',linestyle='--',label='raw')
plt.grid()
plt.legend()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude of driven filtered')
plt.savefig('plot9b.png')
plt.grid()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude of driven (cm)')
plt.savefig('plot10b.png')
```
