Loading the data using Pandas a Python Program which allows us to read spreadsheets

```python
import pandas
datav = pandas.read_excel("Interferoemter Data (2).xlsx",sheet_name="Sheet4")
#datav[0:5] test to see if working
```

In order for us to analyse this data we need to plot a graph of Distance Vs Voltage

```python
#Importing the python packages to allow us to plot a graph
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

plt.plot(datav['Distance mm'],datav['volt 1'],color='blue',label='Volt 1',linestyle='-',)
plt.plot(datav['Distance mm'],datav['volt 2'],linestyle='-',color='red',label='Volt 2',)
plt.plot(datav['Distance mm'],datav['volt 3'],linestyle='-',color='green',label='Volt 3',)
plt.plot(datav['Distance mm'],datav['volt 4'],linestyle='-',color='yellow',label='Volt 4',)
plt.plot(datav['Distance mm'],datav['volt 5'],linestyle='-',color='black',label='Volt 5',)
plt.legend()
plt.xlabel('Distance (mm)')
plt.ylabel('Voltage (V)')
plt.title('A graph to show voltage vs distance')
plt.savefig('Plot of distance vs voltage.png')
plt.grid()
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 6)
fig.savefig('Plot of distance vs voltage.png', dpi=100)
```

As we can see the yellow and black lines are out of sync in the spikes at 118,132,146.
Lets plot agin without Volts 4 and 5

```python
plt.plot(datav['Distance mm'],datav['volt 1'],color='blue',label='Volt 1',linestyle='-',)
plt.plot(datav['Distance mm'],datav['volt 2'],linestyle='-',color='red',label='Volt 2',)
plt.plot(datav['Distance mm'],datav['volt 3'],linestyle='-',color='green',label='Volt 3',)
plt.legend()
plt.xlabel('Distance (mm)')
plt.ylabel('Voltage (V)')
plt.title('A Graph to show distance vs voltage (which all follow the trend)')
plt.savefig('A plot of distance vs voltage2.png')
plt.grid()
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 6)
fig.savefig('A plot of distance vs voltage2.png', dpi=100)
```

With this second graph we can better see the pattern.
Therefore lets calculate an average with volts 1,2,3

```python
datav['V avg']=(datav['volt 1']+datav['volt 2']+datav['volt 3'])/3

## as a test, print the data table
#print(data)
```

Now we have our average volt readings from volt 1,2 and 3
lets plot the graph


```python
plt.plot(datav['Distance mm'],datav['V avg'],color='blue',label='V avg',linestyle='-',)
plt.legend()
plt.xlabel('Distance (mm)')
plt.ylabel('Average Voltage (V)')
plt.title('A graph to show distance vs avg voltage')
plt.savefig('plot of distance vs avg volt.png')
plt.grid()
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 6)
fig.savefig('plot of distance vs avg volt.png', dpi=100)
```

Lets see how well the average agrees with volts 1 2 and 3

```python
plt.plot(datav['Distance mm'],datav['volt 1'],color='yellow',label='Volt 1',linestyle='',marker='.')
plt.plot(datav['Distance mm'],datav['volt 2'],linestyle='',color='red',label='Volt 2',marker='.')
plt.plot(datav['Distance mm'],datav['volt 3'],linestyle='',color='green',label='Volt 3',marker='.')
plt.plot(datav['Distance mm'],datav['V avg'],color='blue',label='V avg',linestyle='-',)
plt.legend()
plt.xlabel('Distance (mm)')
plt.ylabel('Voltage (V)')
plt.title('A graph to show Distance vs avg volt and the volts which follow the trend')
plt.savefig('plot of Distance vs avg volt and volts which follow the trend.png')
plt.grid()
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 6)
fig.savefig('plot of Distance vs avg volt and volts which follow the trend.png', dpi=100)
```

As can been seen the readings volt 1,2 and 3 agree with the average voltage quiet well
Now lets have a look at distance warped.
