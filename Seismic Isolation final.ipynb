Below I have imported Pandas in order to read the spreadhseet

```python
import pandas 
data = pandas.read_excel("Sesmic Isolation data final.xlsx",sheet_name="Sheet1")
#data[0:10]
```

Below I am Importing the nescarry packages for graph ploting and ploting frequency vs Horizontal amplitude of driven mirror

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(data['Frequency (Hz)'],data['Horizontal amplitude of driven mirror (cm)'],color='red',linestyle='-')
plt.grid()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Horizontal amplitude of a driven mirror (cm)')
plt.savefig('plot of frequency vs Horizontal amplitude of driven mirror.png')
plt.title('A graph of Frequency vs Horizontal amplitude of a driven mirror, Fig 4.1')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 6)
fig.savefig('plot of frequency vs Horizontal amplitude of driven mirror.png', dpi=100)
```

Below I am doing two things firstly ploting the same graph as above and caulcuating and ploting a regression line for the line.

```python
#Importing stats packages
from statistics import mean
import numpy as np

xs = data['Frequency (Hz)']
ys = data['Horizontal amplitude of driven mirror (cm)']

def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    
    b = mean(ys) - m*mean(xs)
    
    return m, b

m, b = best_fit_slope_and_intercept(xs,ys)

regression_line = [(m*x)+b for x in xs]
import matplotlib.pyplot as plt
plt.plot(data['Frequency (Hz)'],data['Horizontal amplitude of driven mirror (cm)'],color='red',linestyle='-')
plt.plot(regression_line)
plt.grid()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Horizontal amplitude of a driven mirror (cm)')
plt.savefig('plot of frequency vs Horizontal amplitude of driven mirror with regression line.png')
plt.title('A graph of Frequency vs Horizontal amplitude of a driven mirror with regression line')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 6)
fig.savefig('plot of frequency vs Horizontal amplitude of driven mirror with regression line.png', dpi=100)
print("y="+str(b), "x*"+str(m))
```

As we can see there is damping taking place, evident by the fact our regression line y=-0.005x+0.17 has a negative gradient.
