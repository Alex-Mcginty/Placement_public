### Read in data
Use pandas package to read in data

```python
import pandas 
data = pandas.read_excel("Interferoemter Data (1).xlsx",sheet_name="Sheet2")
```

### Plot data

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(data['Distance mm'],data['volt 1'],color='blue',label='V1',linestyle='-',)
plt.plot(data['Distance mm'],data['volt 2'],linestyle='-',color='red',label='V2',)
plt.plot(data['Distance mm'],data['volt 3'],linestyle='-',color='green',label='V3',)
plt.plot(data['Distance mm'],data['volt 4'],linestyle='-',color='yellow',label='V4',)
plt.plot(data['Distance mm'],data['volt 5'],linestyle='-',color='black',label='V5',)
plt.legend()
plt.xlabel('Distance (mm)')
plt.ylabel('Voltage (V)')
plt.savefig('plot1.png')
```

```python
data['V avg']=(data['volt 1']+data['volt 2'])/2
```

```python
plt.plot(data['Distance mm'],data['volt 1'],color='blue',label='V1',linestyle='',marker='.')
plt.plot(data['Distance mm'],data['volt 2'],linestyle='',color='red',label='V2',marker='.')
plt.plot(data['Distance mm'],data['volt 3'],linestyle='',color='green',label='V3',marker='.')
plt.plot(data['Distance mm'],data['volt 4'],linestyle='',color='yellow',label='V4',marker='.')
plt.plot(data['Distance mm'],data['volt 5'],linestyle='',color='black',label='V5',marker='.')
plt.plot(data['Distance mm'],data['avg volt'],color='blue',label='Avgerage Voltage',linestyle='-',)
plt.legend()
plt.xlabel('Distance (mm)')
plt.ylabel('Average Voltage (V)')
```
