Importing pandas and reading the spreadsheet

```python
import pandas
data = pandas.read_excel("Sesmic Isolation Data 1.xlsx",sheet_name="Sheet1")
data
```

Ploting freq vs driven mirror

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(data['Frequency (Hz)'],data['Horizontal amplitude of driven mirror (cm)'],color='red',linestyle='-')
plt.grid()
plt.xlabel('Frequency(Hz)')
plt.ylabel('Horizontal amplitude (cm)')
plt.savefig('plot1b.png')
```

```python
plt.plot(data['Frequency (Hz)'],data['Horizontal amplitude of oscillating mirror (cm)'],color='blue',linestyle='-')
plt.grid()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Horizontal Amplitude (cm)')
plt.savefig('plot2b.png')
```

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(data['Frequency (Hz)'],data['Ratio of oscillating mirror to driven mirror'],color='blue',linestyle='-')
plt.grid()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Ratio')
plt.savefig('plot3b.png')
```
